```python
#importing libraries
import sys
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv('Churn_modelling.csv')
```

```python
df.head()
```

```python
df.tail()
```

```python
#removing unrelated columns
X=df.iloc[:,3:13].values
print(X)
```

```python
y=df.iloc[:,13].values
print(y)
```

```python
from sklearn.preprocessing import LabelEncoder,OneHotEncoder
lb=LabelEncoder()
X[:,2]=lb.fit_transform(X[:,2])
print(X)
```

```python
from sklearn.compose import ColumnTransformer
ct = ColumnTransformer([('ohe', OneHotEncoder(), [1])], remainder='passthrough')
X = np.array(ct.fit_transform(X)) #.toarray()
print(X)
X = X[:, 1:]
print(X)
```

```python
#Splitting the dataset
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2,random_state=0)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
print(X_train)
```

```python
from tensorflow.keras.models import Sequential
from keras.layers import Dense
```

```python
clas=Sequential()
clas.add(Dense(units=6,kernel_initializer='uniform',activation='relu',input_dim=11))
```

```python
clas.add(Dense(units=6,kernel_initializer='uniform',activation='relu'))
```

```python
clas.add(Dense(units=1,kernel_initializer='uniform',activation='sigmoid'))
```

```python
clas.compile(optimizer='adam',loss='binary_crossentropy',metrics=['accuracy'])
```

```python
clas.fit(X_train , y_train , batch_size=10 , epochs=100)
```

```python
y_pred=clas.predict(X_test)
y_pred=(y_pred>0.5)
```

```python
from sklearn.metrics import confusion_matrix,accuracy_score
cm=confusion_matrix(y_pred,y_test)
print(cm)
accuracy_score(y_pred,y_test)
```
